## RFM analysis (recency, frequency, monetary)

```python
### 1.Create a seperate datafrom for analysing recency, frequency, monetary values
rfm_df = Customer_df[['customer_id', 'rencency', 'number_of_orders','profit']]

rfm_df['r_rank'] = rfm_df['rencency'].rank(ascending=False)
rfm_df['f_rank'] = rfm_df['number_of_orders'].rank(ascending=True)
rfm_df['m_rank'] = rfm_df['profit'].rank(ascending=True)
    
    # Normalizing the rank of the customers
rfm_df['r_rank_norm'] = (rfm_df['r_rank']/rfm_df['r_rank'].max())*100
rfm_df['f_rank_norm'] = (rfm_df['f_rank']/rfm_df['f_rank'].max())*100
rfm_df['m_rank_norm'] = (rfm_df['m_rank']/rfm_df['m_rank'].max())*100

rfm_df.drop(columns = ['r_rank', 'f_rank', 'm_rank'], inplace = True)
rfm_df.head()
```

```python
###2.Calculating RFM score
### ( scale of 5, formula: 0.15*Recency + 0.28 * Frequency + 0.57*Monetary)
rfm_df['rfm_score'] = 0.15 * rfm_df['r_rank_norm'] + 0.28 * rfm_df['f_rank_norm'] + 0.57 * rfm_df['m_rank_norm']
rfm_df['rfm_score'] *= 0.05
rfm_df=rfm_df.round(2)

rfm_df[['customer_id', 'rfm_score']].head()
```

```python
### 3. rating customers based on the rfm_score
rfm_df['customer_segment'] = np.where(rfm_df['rfm_score'] > 4.5, 'Top Customer', 
                                      (np.where(rfm_df['rfm_score'] > 4, 'High Value Customer',
                                                (np.where(rfm_df['rfm_score'] > 3, 'Medium Value Customer',
                                                          (np.where(rfm_df['rfm_score'] > 1.6, 'Low Value Customer', 'Lost Customer'))
)))))

rfm_df[['customer_id','rfm_score','customer_segment']].head(20)
```

```python
### visualisation for the customer segmentation
fig = plt.pie(rfm_df.customer_segment.value_counts(), labels = rfm_df.customer_segment.value_counts().index, autopct = '%.0f%%')
plt.title('Customer Segment', fontsize = 12)
plt.set_cmap('winter')

plt.savefig('Customer Segment.jpeg')
```

# Calculating time offset in month as cohort index
the time offset for each transaction allows to evaluate the metrics for each cohort in a comparable fashion

```python
## 1. create 6 variables that capture the integer value of years, months, and days for Transaction and cohort date using the get_date_int() function
def get_date_int(df, column):
    year = df[column].dt.year
    month = df[column].dt.month
    day = df[column].dt.day
    return year, month, day

# Getting the integers for date parts from the `transaction_month` column
transaction_year, transaction_month, _ = get_date_int(Cohort_df, 'transaction_Month')

# Getting the integers for date parts from the `cohortday` column
cohort_year, cohort_month, _ = get_date_int(Cohort_df, 'cohort_month')
```

Now we will calculate the difference between the Invoice Dates and Cohort dates in years, months separately. then calculate the total Months difference between the two. This will be our month’s offset or cohort Index, which we will use in the next section to calculate the retention rate.

```python
## 2.calculate the diff to create index
# get the diff in years
years_diff = transaction_year - cohort_year
# get the diff in months
months_diff = transaction_month - cohort_month

Cohort_df['cohortindex'] = years_diff *12 + months_diff +1  # "+1" for the 1st month is marked as "1"
Cohort_df.head()
```

Here, at first, we create a group() object with CohortMonth and CohortIndex and store it as a grouping.

Then, we call this object, select the customer_id column and calculate the average.

Then we store the results as cohort_data. Then, we reset the index before calling the pivot function to be able to access the columns now stored as indices.

Finally, we create a pivot table bypassing

cohort_month to the index parameter,
cohortindex to the columns parameter,
customer_id to the values parameter.

```python
# counting daily active user from each chort
# counting no.of unique customer Id's falling in each group of cohort_month and cohortindex
cohort_data = Cohort_df.groupby(['cohort_month', 'cohortindex']).nunique()['customer_id'].reset_index()

cohort_data.head()
```

```python
 ## 3.Calculate business metrics: Retension rate
 # calculate the retension count for each cohort_month paired with cohort Index
 # then calculate retension rate for each cohort_month

# create a pivot table 
cohort_data.pivot(index = 'cohort_month', columns = 'cohortindex', values = 'customer_id')

cohort_counts = cohort_data.pivot(index = 'cohort_month', columns = 'cohortindex', values = 'customer_id')
cohort_sizes = cohort_counts.iloc[:, 0]
retention = cohort_counts.divide(cohort_sizes, axis = 0)

retention.round(3) * 100
```

the retention rate dataframe represents Customer retained across Cohorts. We can read it as follows:

Index value represents the Cohort Columns represent the number of months since the current Cohort For instance: The value at CohortMonth 2017-01-01, CohortIndex 3 is 36.1 and represents 36.1% of customers from cohort 2017-01 were retained in the 3rd Month. Also, you can see from the retention Rate DataFrame:

Retention Rate 1st index i.e 1st month is 100% as all the customers for that particular customer signed up in 1st Month
The retention rate may increase or decrease in subsequent Indexes.
Values towards the bottom right have a lot of NaN values.

```python
retention = retention.round(3) *100
retention.index = retention.index.strftime('%Y-%m')

# Visualizing the retention rate
plt.figure(figsize = (16,10))
# adding a title
plt.title('Retention Rates', fontsize = 12)
# creating the heatmap
sns.heatmap(retention, annot = True, vmin = 0.0, vmax=100, cmap="YlGnBu", fmt='g')

plt.yticks(rotation = '360')

plt.show()
```
